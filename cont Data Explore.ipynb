```python
no_years_one = df_one[df_one["txnYear"].astype(str) == "2019"]
```

```python
no_years_one.head()
```

```python
no_years_two = df_one[df_one["txnYear"].astype(str) == "2020"]
```

```python
no_years_two.head()
```

```python
no_years_one.info()
```

```python
needed_col = no_years_one.columns.to_list()
needed_col
```

```python
needed_col = no_years_one.columns.to_list()[6]
needed_col
```

```python
grouped_mnt = pd.DataFrame(no_years_one.groupby(by = "Month")[needed_col].sum())
grouped_mnt
```

```python
grouped_mnt.reset_index(inplace = True)
```

```python
grouped_mnt
```

```python
grouped_mnt["Month"] = grouped_mnt["Month"].apply(lambda a : dt.strptime(a, "%B"))
```

```python
grouped_mnt
```

```python
grouped_mnt.sort_values(by = "Month", inplace = True)
```

```python
grouped_mnt
```

```python
grouped_mnt["Month"] = grouped_mnt["Month"].apply(lambda a : a.strftime("%B"))
```

```python
grouped_mnt
```

```python
X_axis = grouped_mnt["Month"]
```

```python
Y_axis = grouped_mnt["no of txns"]
```

```python
plt.rcParams["figure.figsize"] = [10,9]
```

```python
plt.plot(X_axis, Y_axis, linewidth = 2, alpha = 0.75, color = "red")

plt.title("monthly sales trend 2019")
plt.xlabel("Month")
plt.ylabel("no of txns")

plt.show()
```
